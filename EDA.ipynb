### Join milege and disengagement data 

```python
disengagements_data_2021_22_monthly = disengagements_data_2021_22.groupby(['manufacturer','month_id','vin_number'],as_index = False)['date'].count()
disengagements_data_2021_22_monthly = disengagements_data_2021_22_monthly.rename(columns={'date': 'disengagements'})
```

```python
#disengagements_data_2021_22_monthly.columns
disengagements_data_2021_22_monthly
```

```python
milege_disengagements_data_2021_22_l = pd.merge(milege_data_2021_22_l, disengagements_data_2021_22_monthly,  how='left', left_on=['manufacturer','month_id','vin_number'], right_on = ['manufacturer','month_id','vin_number'])
milege_disengagements_data_2021_22_l['disengagements'] = milege_disengagements_data_2021_22_l['disengagements'].fillna(0)
milege_disengagements_data_2021_22_l['miles'] = milege_disengagements_data_2021_22_l['miles'].fillna(0)
print(milege_disengagements_data_2021_22_l.info())
print(milege_disengagements_data_2021_22_l.head())
```

```python
# Cumulative sum
milege_disengagements_data_2021_22_l=milege_disengagements_data_2021_22_l.sort_values(by=['manufacturer','month_id','vin_number'], ascending = [True, True,True])
milege_disengagements_data_2021_22_l['cumsum_mile']=milege_disengagements_data_2021_22_l.groupby(['manufacturer'])['miles'].cumsum()
milege_disengagements_data_2021_22_l['cumsum_disengagements']=milege_disengagements_data_2021_22_l.groupby(['manufacturer'])['disengagements'].cumsum()
milege_disengagements_data_2021_22_l.head(100)
```
